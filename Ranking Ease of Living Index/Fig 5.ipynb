```python
import pandas as pd
import matplotlib.pyplot as plt

file_path = 'All_Index_ranking.csv'
data = pd.read_csv(file_path)

selected_countries = ["India", "United States", "China", "Australia", "Malaysia", "Canada", "Germany", "United Kingdom"]

highlighted_countries = {
    "United States": "blue", 
    "India": "orange", 
    "China": "red", 
    "Malaysia": "black"
}
bright_colors = ["lime", "fuchsia","yellow", "deepskyblue"]

subset1 = data[(data['Year'] >= 1970) & (data['Year'] <= 1993)]

plt.figure(figsize=(10, 6))
color_index = 0

for country in selected_countries:
    df_country = subset1[subset1['Country'] == country]
    if country in highlighted_countries:
        plt.plot(df_country['Year'], df_country['Quality_Of_Life_Index_FA'], 
                 label=country, linewidth=2.5, color=highlighted_countries[country])

        latest_year = df_country['Year'].iloc[-1]
        latest_value = df_country['Quality_Of_Life_Index_FA'].iloc[-1]
        plt.text(latest_year + 0.1, latest_value, f'{latest_value:.2f}', fontsize=10,
                 color=highlighted_countries[country], va='center')

    else:
        plt.plot(df_country['Year'], df_country['Quality_Of_Life_Index_FA'], 
                 label=country, linewidth=1.5, linestyle='--', color=bright_colors[color_index])
        color_index = (color_index + 1) % len(bright_colors)

plt.xlabel('Year', fontsize=12)
plt.ylabel('Quality of Life Index', fontsize=12)
plt.title('Quality of Life Index (1970–1993)', fontsize=14, weight='bold', ha='left', x=0)
plt.xticks(range(1970, 1995, 5))
plt.legend(title="Country", loc='upper left')
plt.tight_layout()
plt.show()
```
