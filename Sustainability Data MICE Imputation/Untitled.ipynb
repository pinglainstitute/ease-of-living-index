```python
import matplotlib.pyplot as plt

countries = ["Malaysia", "China", "India", "Canada", "Australia",
             "Japan", "United States", "Germany", "United Kingdom"]

emissions = [115.76, 104.05, 81.90, 74.58, 41.32, 35.30, 32.16, 16.58, 13.77]

plt.figure(figsize=(12, 7), facecolor='#f2f2f2')
ax = plt.gca()
ax.set_facecolor('#f2f2f2')

bars = plt.barh(countries, emissions,
                color='#8fd694', edgecolor='black', height=0.6)

# Title + subtitle
plt.suptitle('Greenhouse Gas Emission', fontsize=14, weight='bold', ha='left', x=0.175)
plt.text(0.0, 1.02, "The emission is in metric tonnes.", fontsize=12, style='italic', ha='left', transform=plt.gca().transAxes)

# Axis labels (bigger)
plt.xlabel("Emission", fontsize=18)
plt.ylabel("Country", fontsize=18)

# Tick labels
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)

# Value labels
for bar in bars:
    width = bar.get_width()
    plt.text(width - 1,
             bar.get_y() + bar.get_height()/2,
             f'{width:.2f}',
             va='center',weight='bold', ha='right', fontsize=10, color='black', bbox=dict(facecolor='white', edgecolor='black', boxstyle='round,pad=0.3'))

plt.xlim(0, 125)

# IMPORTANT: use this instead of tight_layout()
plt.subplots_adjust(top=0.85, left=0.18)
plt.savefig('Green House emission_Fig14.png')

plt.show()
```
