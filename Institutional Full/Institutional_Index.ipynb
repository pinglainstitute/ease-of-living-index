```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('Rule of Law 1996-2022.csv')
```

```python
df.columns = [col.split(' ')[0] if 'YR' in col else col for col in df.columns]
```

```python
df.columns = [col.split(' ')[0] if 'YR' in col else col for col in df.columns]
available_years = [col for col in df.columns if col.isdigit()]
series_of_interest = [
    'Control of Corruption: Estimate',
    'Government Effectiveness: Estimate',
    'Political Stability and Absence of Violence/Terrorism: Estimate',
    'Regulatory Quality: Estimate',
    'Rule of Law: Estimate',
    'Voice and Accountability: Estimate'
]

melted_dfs = {}

for series in series_of_interest:

    df_filtered = df[df['Series Name'] == series]
    
    melted = pd.melt(df_filtered, 
                     id_vars=['Country Name'], 
                     value_vars=available_years, 
                     var_name='Year', 
                     value_name=series)
    
    melted = melted.dropna(subset=[series])
    
    melted['Year'] = melted['Year'].astype(int)
    
    melted_dfs[series] = melted


merged_df = melted_dfs[series_of_interest[0]]

for series in series_of_interest[1:]:
    merged_df = pd.merge(merged_df, melted_dfs[series], on=['Country Name', 'Year'], how='outer')

merged_df.reset_index(drop=True, inplace=True)
merged_df.to_csv('Final_Merged_Series_Data.csv', index=False)

merged_df.head()
```
