```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Fig 7

# Load the data
file_path = 'GDP_Growth_Rate_Final.csv'
data = pd.read_csv(file_path)

# Define selected countries and years
selected_countries = ["India", "United States", "China", "Australia", "Malaysia", "Canada", "Germany", "United Kingdom", "Japan"]
years = list(range(2013, 2024))

# Filter the data
gdp_data = data[
    (data['Country'].isin(selected_countries)) &
    (data['Year'].isin(years))
][['Country', 'Year', 'GDP_Growth_Rate']]  # Ensure column name matches exactly

# Calculate average GDP growth rate for each country
average_gdp = gdp_data.groupby('Country')['GDP_Growth_Rate'].mean().reset_index()
average_gdp.rename(columns={'GDP_Growth_Rate': 'Average_GDP_Growth_Rate'}, inplace=True)
average_gdp_sorted = average_gdp.sort_values(by='Average_GDP_Growth_Rate', ascending=True)

plt.figure(figsize=(10, 6))

# Horizontal bar plot
plt.bar(
    average_gdp_sorted['Country'],    
    average_gdp_sorted['Average_GDP_Growth_Rate'],
    color='skyblue',
    edgecolor='black',
    width=0.7
)
plt.rcParams.update({
    'font.size': 14,
    'axes.titlesize': 16,
    'axes.labelsize': 14,
    'xtick.labelsize': 14,   # increase this
    'ytick.labelsize': 14,   # increase this
    'legend.fontsize': 12
})

# Titles
plt.suptitle('GDP Growth Rate', fontsize=14, weight='bold', ha='left', x=0.13)
plt.text(0.0, 1.02, "Average GDP Growth Rate over the past 10 years", fontsize=12, style='italic', ha='left', transform=plt.gca().transAxes)

plt.xlabel('Country', fontsize=12)
plt.ylabel('Average GDP Growth Rate (%)', fontsize=12)

# Adding text labels inside bars
for index, value in enumerate(average_gdp_sorted['Average_GDP_Growth_Rate']):
    plt.text(index, value - 0.5, f'{value:.2f}%', ha='center', va='bottom', fontsize=10, weight='bold', color='black', bbox=dict(facecolor='white', edgecolor='black', boxstyle='round,pad=0.3'))

plt.xticks(rotation=45, ha='right') 
plt.tight_layout()
plt.savefig('GDP growth rate.png')
plt.show()
```

```python
# FIG 7 – GDP Growth Rate (Average 2013–2023)

import pandas as pd
import matplotlib.pyplot as plt

# ---------------------------
# Font Configuration (Medium Style)
# ---------------------------
plt.rcParams.update({
    'font.size': 12,
    'axes.titlesize': 16,
    'axes.labelsize': 14,
    'xtick.labelsize': 12,
    'ytick.labelsize': 12,
    'legend.fontsize': 12
})

# ---------------------------
# Load Data
# ---------------------------
file_path = 'GDP_Growth_Rate_Final.csv'
data = pd.read_csv(file_path)

selected_countries = [
    "India", "United States", "China", "Australia",
    "Malaysia", "Canada", "Germany",
    "United Kingdom", "Japan"
]

years = list(range(2013, 2024))

gdp_data = data[
    (data['Country'].isin(selected_countries)) &
    (data['Year'].isin(years))
][['Country', 'Year', 'GDP_Growth_Rate']]

# ---------------------------
# Compute Average Growth
# ---------------------------
average_gdp = (
    gdp_data.groupby('Country')['GDP_Growth_Rate']
    .mean()
    .reset_index()
)

average_gdp.rename(
    columns={'GDP_Growth_Rate': 'Average_GDP_Growth_Rate'},
    inplace=True
)

average_gdp_sorted = average_gdp.sort_values(
    by='Average_GDP_Growth_Rate',
    ascending=True
)

# ---------------------------
# Plot
# ---------------------------
plt.figure(figsize=(10, 6))

plt.bar(
    average_gdp_sorted['Country'],
    average_gdp_sorted['Average_GDP_Growth_Rate'],
    color='skyblue',
    edgecolor='black',
    width=0.7
)

# Titles
plt.title('GDP Growth Rate', weight='bold', ha='left', x=0.13)

plt.xlabel('Country')
plt.ylabel('Average GDP Growth Rate (%)')

# ---------------------------
# Medium Annotation Size
# ---------------------------
for index, value in enumerate(average_gdp_sorted['Average_GDP_Growth_Rate']):
    plt.text(
        index,
        value + 0.01,
        f'{value:.2f}%',
        ha='center',
        va='bottom',
        fontsize=14,      # Medium annotation size
        weight='bold',
        color='black'
    )

plt.xticks(rotation=45, ha='right')
plt.tight_layout()

plt.savefig('GDP_growth_rate.png', dpi=300, bbox_inches='tight')
plt.show()
```

```python
#Fig 8

import pandas as pd
import matplotlib.pyplot as plt

file_path = 'Cost of Living.csv'  # Change to your actual file path
data = pd.read_csv(file_path)

year = 2021
selected_countries = ["India", "United States", "China", "Australia", "Malaysia", "Canada", "Germany", "United Kingdom", "Japan"]

# Filter all countries for year 2021
data_2021 = data[data['Year'] == year]

data_2021['Scaled_Index'] = (
    data_2021['Cost of Living Index'] - data_2021['Cost of Living Index'].min()
) / (
    data_2021['Cost of Living Index'].max() - data_2021['Cost of Living Index'].min()
)

filtered_data = data_2021[data_2021['Country'].isin(selected_countries)].copy()

plt.figure(figsize=(10, 6))

plt.barh(
    filtered_data['Country'],
    filtered_data['Scaled_Index'],
    color='#FF7F50',
    edgecolor='black'
)

plt.suptitle('Cost of Living Index', fontsize=14, weight='bold', ha='left', x=0.2)
plt.text(0.0, 1.02, "The Cost of Living Index is scaled between 0 and 1", fontsize=12, style='italic', ha='left', transform=plt.gca().transAxes)

plt.xlabel('Cost of Living Index', fontsize=14)
plt.ylabel('Country', fontsize=14)

plt.xticks(fontsize=13)
plt.yticks(fontsize=13)
# Adding text labels inside bars
for index, value in enumerate(filtered_data['Scaled_Index']):
    plt.text(value-0.005, index, f'{value:.2f}', va='center',weight='bold', ha='right', fontsize=10, color='black', bbox=dict(facecolor='white', edgecolor='black', boxstyle='round,pad=0.3'))

plt.tight_layout()
plt.savefig('Cost of living index.png')
plt.show()
```

```python
# Fig 8

import pandas as pd
import matplotlib.pyplot as plt

# ---------------------------
# Font Configuration (Medium Style)
# ---------------------------
plt.rcParams.update({
    'font.size': 12,
    'axes.titlesize': 16,
    'axes.labelsize': 14,
    'xtick.labelsize': 12,
    'ytick.labelsize': 12
})

file_path = 'Cost of Living.csv'
data = pd.read_csv(file_path)

year = 2021
selected_countries = [
    "India", "United States", "China", "Australia",
    "Malaysia", "Canada", "Germany",
    "United Kingdom", "Japan"
]

data_2021 = data[data['Year'] == year].copy()

# Scale between 0 and 1
data_2021['Scaled_Index'] = (
    data_2021['Cost of Living Index'] - data_2021['Cost of Living Index'].min()
) / (
    data_2021['Cost of Living Index'].max() - data_2021['Cost of Living Index'].min()
)

filtered_data = data_2021[data_2021['Country'].isin(selected_countries)].copy()

plt.figure(figsize=(10, 6))

plt.barh(
    filtered_data['Country'],
    filtered_data['Scaled_Index'],
    color='#FF7F50',
    edgecolor='black'
)

plt.title('Cost of Living Index', weight='bold', ha='left', x=0.13)

# ---------------------------
# Medium Annotation Size (14)
# ---------------------------
for index, value in enumerate(filtered_data['Scaled_Index']):
    plt.text(value+0.005, index, f'{value:.2f}', va='center',weight='bold', ha='right', fontsize=11, color='black', bbox=dict(facecolor='white', edgecolor='black', boxstyle='round,pad=0.3'))

plt.xlabel('Scaled Index (0–1)')
plt.tight_layout()
plt.savefig('cost_of_living.png', dpi=300, bbox_inches='tight')
plt.show()
```
